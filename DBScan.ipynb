```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
data = pd.read_csv("./BLE_RSSI_dataset/iBeacon_RSSI_UnLabeled.csv")
data.head()
#drop Time Location
newdata = data.drop(columns=['location','date']);
newdata.head()
```

```python
train_data=newdata[['b3001','b3002','b3003','b3004','b3005','b3006','b3007','b3008','b3009','b3010','b3011','b3012','b3013',]]
db = DBSCAN(eps=20, min_samples=80).fit(train_data)
labels = db.labels_ 
newdata['cluster_db'] = labels  # 在数据集最后一列加上经过DBSCAN聚类后的结果
newdata.sort_values('cluster_db')
```

```python
#用TSNE进行数据降维并展示聚类结果
from sklearn.manifold import TSNE
tsne = TSNE()
tsne.fit_transform(newdata) #进行数据降维,并返回结果
tsne = pd.DataFrame(tsne.embedding_, index = data.index) #转换数据格式

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号

#不同类别用不同颜色和样式绘图
d = tsne[newdata['cluster_db'] == 0]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 1]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 2]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 3]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 4]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 5]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 6]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 7]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 8]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 9]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 10]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 11]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 12]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 13]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 14]
plt.plot(d[0], d[1],'.')
d = tsne[newdata['cluster_db'] == 15]
plt.plot(d[0], d[1],'.')

plt.show()
```
